### Load data

```python
import pandas as pd

df = pd.read_feather('./experiment/data_to_analysis/scores_with_dimension_compressed.ftr')
df
```

```python
from matplotlib import pyplot as plt

models = df.columns[-7:]

winner = df[df['win']==True].sample(20000)
loser = df[df['win']==False].sample(20000)

x_win = winner['pca'].to_numpy()
y_wins = [winner[model].to_numpy() for model in models]
x_lose = loser['pca'].to_numpy()
y_loses = [loser[model].to_numpy() for model in models]
```

### Draw graphs

```python
i = 0 # Model 2
tmax = max(max(y_wins[i]), max(y_loses[i]))
tmin = min(min(y_wins[i]), min(y_loses[i]))
plt.figure(figsize=(5, 3))
plt.yticks((tmin, tmax))
plt.xlabel('Dimension reduced feature value')
plt.ylabel('Model score')
plt.scatter(x_win, y_wins[i], color=(0,0,1), s=8, alpha=0.2, label='winner')
plt.scatter(x_lose, y_loses[i], color=(1,0,0), s=8, alpha=0.2, label='loser')
plt.title('Model 2')
plt.legend(loc='upper right', labelcolor=('b', 'r'), ncol=2, handlelength=0)
plt.show()
```
